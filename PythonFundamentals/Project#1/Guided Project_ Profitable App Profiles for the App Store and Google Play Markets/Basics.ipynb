# Duplicate Entries
Sometimes, we may find duplicate entries inside of different data sets, for example, in the Google Play Data Set, we find duplicates of the Instagram app. We can use a 'for' loop to go through the data set at hand, and find duplicates.

```python
for app in android:
    name  = app[0]
    if name  == 'Instagram': #If the first element is 'Instagram' display row
        print(app)
```

Note here, that there are 4 (four) <i>different</i> rows with duplicate apps. 

We can expand on this, and find the total amount of duplicate apps in the data by declareing two new lists, and filling them with the information found in each of the rows:

```python
duplicate_apps = [] #Holds Duplicate apps in set
unique_apps = [] #Holds all unique apps in set

for app in android:
    name = app[0] #assign the app name to the first elem.
    if name in unique_apps: #if we find the duplicate, append. else - append to other list
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)

#Show number of duplicate apps, and examples of such apps:
print('Number of duplicate apps: ', len(duplicate_apps))
print('\n')
print('Examples of duplicate apps: ', duplicate_apps[:7])
```

As we can see above, by createing two new empty lists, we are able to find all of the duplicate apps that are located within the data list with little trouble by iterating through it useing a simple for loop.

Useing the len() function on the list gives us a simple way of finding all of the duplicates.

If you examine the rows we printed two cells above for the Instagram app, the main difference happens on the fourth position of each row, which corresponds to the number of reviews. The different numbers show that the data was collected at different times. We can use this to build a criterion for keeping rows. We won't remove rows randomly, but rather we'll keep the rows that have the highest number of reviews because the higher the number of reviews, the more reliable the ratings.

To do that, we will:

*Create a dictionary where each key is a unique app name, and the value is the highest number of reviews of that app
    
*Use the dictionary to create a new data set, which will have only one entry per app (and we only select the apps with the highest number of reviews)


# Duplicate Entries (Part Two)

As mementined before to facilitate our goal in keeping rows that we want, we will create a dictionary:

```python
reviews_max = {} #Empty dictionary containing maximum reviews

for row in android:
    name = row[0] #name is in the first row:
    n_reviews = float(row[3]) #num of reviews are in index 3:
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max: #otherwise create new entry 
        reviews_max[name] = n_reviews

#Display the number of entries (the length of the new dictionary, reviews_max:
print('The amount of entries are now: ', len(reviews_max))
    
    
```

This is the expected length we wanted to have, 9659 entries.
Now, our next step is to use the same dictionary to help remove duplicate rows for the data set.


```python
android_clean = [] #Stores new cleaned data set:
already_added = [] #Stores app names:

for row in android:
    name  = row[0] #Holds the name of each app
    n_reviews = float(row[3]) #Holds no. of reviews:
    if n_reviews == reviews_max[name] and (name not in already_added):
        android_clean.append(row)
        already_added.append(name)
        
#Print the data set 
print('The amount of data entries should be 9659: ', len(android_clean))
```

So what did we do to obtain the data entries we wanted? 
First, we declared two new, empty lists: 'android_clean' and 'already_added'

We then iterated through the list with a 'for-loop' 
-Assigning the name to the first index in the row (row[0])
-As well as the number of reviews (n_reviews) to a casted float value in index 3 (row[3])

Useing an 'if' statement, we then checked if n_reviews had the same amount as the number of maximum reviews of the app name that is (reviews_max[name]) 'anded' if the name is not alwready in the list

If this is the case, then we will:
<ul>-Append the entire row to the list android_clean 
(via android_clean.append(row))</ul>
<ul>-Append the name of the app to the list already_added 
(via alwready_added.append(name))</ul>

<u>NOTE</u>: The reason for the if statement and the supplementary condition above is because we want the condition to account for cases where the <strong> highest number of reviews of a duplicate app is the <i>same</i> for one entry</strong>. If we just did n_reviews == reviews_max[name], we would still get duplicates later on, which isn't want we wanted.

The english characters used in text all range from 0 - 127 according to the ASCII system. Based on the number system we are also able to distinguish which characters are used in english text and which are not.

In Python, strings are <i>indexable</i>, and <i>iterable</i>, which means that are are able to use indexing to select an individual character, and also iterate on the string using a 'for' loop.

Here is an example:

```python
string = 'abc'

print(string[0])
print(string[1])
print(string[2])

print('\n')
print('The following should also be the same: ')
print('\n')
for character in string:
    print(character)
```

Now that we know the flexibility of Python, lets write a function that will tells us if the string passed as an argument contains characters that are not found in English text:

```python
def notEngChar(string_):
    for char in string_:
        if ord(char) > 127:
            return True
    return False


#Display the result:
print(notEngChar('Instagram'))
print(notEngChar('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(notEngChar('Docs To Go‚Ñ¢ Free Office Suite'))
print(notEngChar('Instachat üòú'))
```

Instead of just checking one character, another option would be checking three characters or less that fall outside the ASCII range:

```python
def explore_data(dataset, start, end, rows_and_columns = False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')  #adds a new (empty) line after each row
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))




def ThreeChars(string_):
    charCount = 0;
    #Use a for loop to iterate through the string,
    #if there is a character beyond 127, increment charCount.
    for char in string_:
        if (ord(char) > 127):
            charCount+=1
    #If the number of char. exceeds 3 return false, otherwise return true
    if charCount > 3:
        return False
    else:
        return True


    
#Some examples of output:
print(ThreeChars('Instachat üòú'))
print(ThreeChars('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(ThreeChars('Docs To Go‚Ñ¢ Free Office Suite'))

android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if ThreeChars(name):
        android_english.append(app)
        
for app in ios:
    name = app[1]
    if ThreeChars(name):
        ios_english.append(app)
        
explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

# Part Two

We'll build two functions we can use to analyze the frequency tables:

<ul>One function to generate frequency tables that show percentages.</ul>
<ul>Another function that we can use to display the percentages in a descending order</ul>


```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total+=1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
            
    table_percentages = {}
    
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage
        
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
    
    
    
```

# Most Popular Apps on Google Play
For the Google Play market, we actually have data about the number of installs, so we should be able to get a clearer picture about genre popularity. However, the install numbers don't seem precise enough ‚Äî we can see that most values are open-ended (100+, 1,000+, 5,000+, etc.):

```python
display_table(android_apps_final, 5) # the Installs columns
```



One problem with this data is that is not precise. For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we don't need very precise data for our purposes ‚Äî we only want to get an idea which app genres attract the most users, and we don't need perfect precision with respect to the number of users.

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on.

To perform computations, however, we'll need to convert each install number to float ‚Äî this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).


```python


categories_android = freq_table(android_apps_final, 1)

for category in categories_android:
    total = 0
    len_category = 0
    for app in android_apps_final:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)

```
